```python
def f(x):

    a = pd.Series([min(i) for i in x], index=x.index)
    b = pd.Series([max(i) for i in x], index=x.index)
    c = pd.Series([sum(i)/len(i) for i in x], index=x.index)
    return pd.concat([a,b,c], keys=('min','max','mean'))
```
